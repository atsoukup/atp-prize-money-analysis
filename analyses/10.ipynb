# Total Prize Money Awarded at Each Tournament Level

The total prize money awarded at each tournament level is calculated by adding up all prize money earned by players from each respective tournament type. The charts show the trend of the amount of prize money earned each year at the various tournament levels.

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
START_YEAR = 2009
END_YEAR = datetime.datetime.now().year - 1
TOURNAMENT_TYPES = ["Grand Slam", "ATP Masters 1000", "ATP 500", "ATP 250", "Challenger"]
```

```python
df = pd.read_csv(f"../data/tournaments/tournaments.gz")
```

```python
singles = df[df["type"] == "singles"]
doubles = df[df["type"] == "doubles"]
```

```python
for tournament_type in TOURNAMENT_TYPES:
    df = singles[singles["tournament_type"] == tournament_type]

    plt.plot(df["year"], df["total_prize_money"], label=tournament_type)
    plt.title(f"Total Prize Money Awarded at the {tournament_type} Tournament Level")
    plt.xlabel("Year")
    plt.ylabel("Prize Money")
    plt.show()

    print(df[["year", "total_prize_money"]])
```
