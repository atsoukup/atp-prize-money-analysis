```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
START_YEAR = 2009
END_YEAR = datetime.datetime.now().year - 1
```

```python
data = []
for year in range(START_YEAR, END_YEAR + 1):
    for t in ["singles", "doubles"]:
        df = pd.read_csv(f"../data/players/clean/{t}/{year}.gz")
        median = df["prize_money"].median()
        average = df["prize_money"].mean()
        data.append({
            "year": year,
            "type": t,
            "median_prize_money": median,
            "average_prize_money": average
        })

result = pd.DataFrame(data)
```

```python
for t in ["singles", "doubles"]:
    df = result[result["type"] == t]
    plt.plot(df["year"], df["median_prize_money"], label="median")
    plt.plot(df["year"], df["average_prize_money"], label="average")
    plt.legend()
    plt.xlabel("Year")
    plt.ylabel("Prize Money")
    plt.title(f"Median and Average Prize Money for {t.capitalize()}")
    plt.show()
```

```python
for t in ["singles", "doubles"]:
    df = result[result["type"] == t]
    plt.plot(df["year"], df["average_prize_money"] / df["median_prize_money"])
    plt.xlabel("Year")
    plt.ylabel("Ratio")
    plt.title(f"Ratio of Average Prize Money to Median Prize Money for {t.capitalize()}")
    plt.show()
```
